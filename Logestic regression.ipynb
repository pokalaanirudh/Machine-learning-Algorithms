```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv("ML/Datasets/Social_Network_Ads.csv")
```

```python
df.head(10)
# two independent variables age and estimated salary 
# our output will be purchased
```

```python
real_x = df[["Age","EstimatedSalary"]].values
real_y = df[["Purchased"]].values
```

```python
train_x,test_x,train_y,test_y = train_test_split(real_x,real_y, test_size=0.25, random_state=0)
```

```python
train_x # here  values of age are in 100's and salary are in 10k too uch difference 
# so we do feat4ure scaling and bring values of independent variables between +2 and -2
```

```python
scaler = StandardScaler()
train_x =scaler.fit_transform(train_x)
test_x = scaler.fit_transform(test_x)
```

```python
classifier_LR = LogisticRegression(random_state=0)
classifier_LR.fit(train_x,train_y)
```

```python
y_pred = classifier_LR.predict(test_x)
```

```python
y_pred
```

```python
test_y
```

```python
cm = confusion_matrix(test_y,y_pred)
```

```python
cm
```

```python
from matplotlib.colors import ListedColormap
x_set ,y_set = train_x, train_y    

x1, x2 = np.meshgrid(np.arange(start = x_set[:,0].min() - 1, stop = x_set[:,0].max() +1 ,step =0.01 ),
                     np.arange(start = x_set[:,1].min() - 1, stop = x_set[:,1].max() +1 ,step =0.01 ))

plt.contourf(x1, x2 , classifier_LR.predict(np.array([x1.ravel(),x2.ravel()]).T).reshape(x1.shape),alpha =0.75, cmap=ListedColormap(("red","blue")))
  
          
plt.xlim(x1.min(),x1.max())
plt.ylim(x2.min(),x2.max())

for i,j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0],x_set[y_set==j, 1], c= ListedColormap(("red","blue"))(i), lable=j)
    
plt.tilte("Logestic regression train")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
